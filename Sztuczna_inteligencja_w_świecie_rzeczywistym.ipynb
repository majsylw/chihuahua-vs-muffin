# Jak uczą się maszyny?

Niektóre technologie sztucznej inteligencji istnieją od dawna, ale postęp w zakresie mocy obliczeniowej, dostępność ogromnych ilości danych i nowe algorytmy doprowadziły do wielkich przełomów w tej dziedzinie. 
Sztuczna inteligencja jest szeroko stosowana do dostarczania spersonalizowanych rekomendacji podczas zakupów, czy zwykłego wyszukiwania informacji w sieci. Do bardziej zaawansowanych wynalazków należą samochody autonomiczne - samojeżdżące - które w dużym uproszczeniu decyzje co do wykonania kolejnych ruchów podejmują na bazie danych zebranych z różnego rodzaju zamontowanych w nich czujników.

## Rodzaje uczenia się maszyn

Uczenie maszynowe możemy postrzegać jako sztukę wydobywania wiedzy z danych. Podstawowy podział dziedziny na podobszary wynika z rodzaju zadania, który maszyna ma rozwiązać:
- uczenie nadzorowana (z nauczycielem): występuje gdy wszystkie dane przedstawiane maszynie są zalabelowane, czyli oznaczone, w dokładnie taki sam sposób, jak odpowiedź, której oczekujemy;
- uczenie nienadzorowane (bez nauczyciela): występuje gdy dysponujemy dużą ilością danych bez oznaczeń, a zadanie maszyny polega na ustaleniu struktury danych, np. zgrupowaniu ich odpowiednio;
- uczenie przez wzmacnianie (z krytykiem): w drodze prób i błędów maszyna poszukuje rozwiązania sformułowanego zadania, za swoje działania jest nagradzana (gdy postępuje słusznie) lub karana (gdy się myli), jednak poza tym nie udzielane są jej żadne wskazówki ani sugestie.

Aby móc zdefiniować sieć neuronową, musimy postawić przed nią dobrze zdefiniowane zadanie - takim zadaniem będzie klasyfikacja zdjęć, a dokładniej będziemy rozpoznawać czy na zdjęciu znajduje się pies rasy chihuahua, czy muffinka.

## Pierwsze kroki z PyTorchem - budujemy sieć neuronową

W ogólności sieć neuronowa ogólnie wygląda tak jak na obrazku poniżej. Wejście do sieci znajduje się po lewej stronie, wyjście po prawej. Liczba neuronów wyjściowych odpowiada liczbie klas.

<img src="obrazy/siec_cls.png" width="700">

Zdefiniujmy podobną architekturę dla naszego klasyfikatora. W naszym przypadku mamy dwie klasy: muffinki i psy rasy chihuahua.

```python
# Zdefinujemy własną sieć bazując na module PyTorcha
class Klasyfikator(nn.Module):
    """
    Klasa definująca sieć neuronową - Klasyfikator
    """
    def __init__(self, input_dim=(3, 224, 224)):
        """
        Podstawową jednostką pozwalająca na zbudowanie własnej architektury w Pytorch jest klasa nn.Module.
        Nasza sieć Klasyfikator dziedziczy po nn.Module
        
        input_dim: krotka reprezentująca wymiary obrazu (liczbę_kanałów, wysokość, szerokość)
        """
        super().__init__()
        # całkowita liczba pikseli składająca się na obraz
        num_in_features = input_dim[0] * input_dim[1] * input_dim[2]
        # warstwa na wejściu do sieci
        self.layer_0 = nn.Linear(num_in_features, 128)
        # ukryte warstwy
        self.layer_1 = nn.Linear(128, 64)
        self.layer_2= nn.Linear(64, 32)
        # warstwa wyjściowa, rozmiar na wyjściu wynosi 2, gdyż mamy 2 klasy: chihuahua i muffin
        self.layer_3= nn.Linear(32, 2)

    def forward(self, x):
        """
        Definujemy metodę forward, która określa sposób,
        w jaki chcemy, aby dane przepływały przez sieć
        """
        batch_size = x.shape[0]
        # zamieniamy naszą tablicę RGB na pojedynczy wektor
        x = x.view(batch_size, -1)
        
        # aktywacja
        x = F.relu(self.layer_0(x))
        x = F.relu(self.layer_1(x))
        x = F.relu(self.layer_2(x))
        x = F.relu(self.layer_3(x))
        
        # zamieniamy surowe wyjście na prawdopodobieństwa
        x = F.softmax(x, dim=1)
        
        return x
```

```python
model = Klasyfikator()                      # tworzymy nowy obiekt - naszą sieć neuronową
model
```

(Tensor to nic innego jak wielowymiarowa macierz, czyli tablica n-k). To jest dobry moment, aby zastanowić się czym właściwie dla komputera jest obraz. Otóż kolorowe obrazy w modelu RGB (czyli red - czerwony, green - zielony, blue - niebieski) posiadają trzy kanały kolorów. Oznacza to tyle, że tak naprawdę składają się z 3 nakładających się na siebie tablic o szerokości (liczba kolumn) i wysokości (liczba wierszy) obrazu. Teraz każdy piksel ma zdefiniowane 3 liczby, które odpowiadają za stopień nasycenia odpowiedniej barwy (czerwonej, zielonej i niebieskiej). Wartości owych liczb mieszczą się w zakresie od 0 do 255. Znaczy to mniej więcej tyle, że aby uzyskać kolor czerwony wartości poszczególnych elementów piksela powinny prezentować się następująco: R - 255, G - 0 i B - 0. Aby przekształcić obraz na tensor wykorzystamy moduł Pytorcha 'dataloader' - „modułu ładujący dane”. Owy moduł pakuje dane pakuje do pól, z których nasz model już potrafi skorzystać. Możecie o tym myśleć w taki sposób, jakby jedna osoba dysponowała wieloma pudełkami (w których mieszczą się obrazy - tensory) drugiej.
<img src = "obrazy/dataloader.jpg" width = "400">

Oprócz tego musimy zdefiniować też pewne transformacje, aby obrazy mogły zostać przekształcone w tensory i zostać odczytane przez naszą sieć (zarówno zbiór treningowy, jak i walidacyjny). Kontynuując powyższy przykład - aby obrazy można było upakować w odpowiednich pudełkach, najpierw muszą zostać odpowiednio zmniejszone/zwiększone czy nawet przycięte.

```python
normalize = transforms.Normalize(mean=[0.5, 0.5, 0.5],
                                 std=[0.5, 0.5, 0.5])

# transformacje zadane dla danych treningowych
train_transforms = transforms.Compose([
    # zmiana rozmiaru obrazka, aby zmieścił się na warstwę wejściową sieci
    transforms.Resize((224,224)),
    # zamieniamy obrazek na tensor Pytorchowy
    transforms.ToTensor(),
    normalize
])

# poniższe transformacje są dokładnie takie same jak dla zbioru treningowego
validation_transforms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    normalize
])

print("Zastosowane transformacje:", train_transforms)
```

Za pomocą metody ImageFolder utworzymy nasz zbiór danych.

```python
image_datasets = {
    'train':
        datasets.ImageFolder('data/train', train_transforms),
    'validation':
        datasets.ImageFolder('data/validation', validation_transforms)}

print("==Zbiór danych treningowych==\n", image_datasets["train"])
print()
print("==Zbiór danych testowych==\n", image_datasets["validation"])
```

```python
# na koniec tworzymy obiekt DataLoader

dataloaders = {
    'train':
        torch.utils.data.DataLoader(
            image_datasets['train'],
            batch_size=8,
            shuffle=True,
            num_workers=4),
    'validation':
        torch.utils.data.DataLoader(
            image_datasets['validation'],
            batch_size=8,
            shuffle=False,
            num_workers=4)}

print("Dane treningowe:", dataloaders["train"])
print("Dane testowe:", dataloaders["validation"])
```

## Trening - uczymy sieć neuronową

Hurra! Zbudowaliśmy sieć neuronową i mamy dane, aby ją wyuczyć. Teraz **wielokrotnie będziemy pokazywać jej dane ze zbioru treningowego, aby wytrenować model.**

```python
import tqdm

def train_model(model, dataloaders, loss_function, optimizer, num_epochs):
    """
    Trenujemy model na podstawie zadanej funkcji strat oraz algorytmu optymalizującego wagi przez pewną liczbę epok
    """
    epoch_loss_list = []
    epoch_acc_list = []
    # trenujemy model przez n epok - epoka to pełen przebieg modelu przez zbiór treningowy (model widzi wszystkie dane n razy)
    for epoch in tqdm.notebook.trange(num_epochs, desc="Total progress", unit="epoch"):
        print('Epoch {}/{}'.format(epoch+1, num_epochs))
        print('----------------')

        # śledzimy średni błąd i dokładność
        running_loss = 0.0
        running_corrects = 0

        # przechodzimy przez każde zdjęcie (i jego etykietę) w zadanym zbiorze
        # (metoda tqdm pozwala na wyświetlenie progresu)
        for inputs, labels in tqdm.notebook.tqdm(dataloaders['train'], unit="batch", leave=False):
            # FORWARD 
            outputs = model(inputs)
            # wyliczamy błąd predykcji
            loss = loss_function(outputs, labels)

            # BACKWARD PASS - niezbędna faza treningu (tutaj model się uczy)
            optimizer.zero_grad()  
            loss.backward()        
            optimizer.step()       # aktualizacja wag

            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            running_corrects += torch.sum(preds == labels.data)

        # wyświetlanie bieżącej dokładności i błędu
        epoch_loss = running_loss / len(image_datasets['train'])
        epoch_acc = running_corrects.double() / len(image_datasets['train'])
        print(f'Trening: błąd: {epoch_loss:.4f}, dokładność: {epoch_acc:.4f} \n')
        epoch_loss_list.append(epoch_loss)
        epoch_acc_list.append(epoch_acc.item())

    return epoch_loss_list, epoch_acc_list
```

Ostatnia rzecz: musimy zdefiniować funkcję, która będzie sieci dawała informację zwrotną na temat tego, 
jak dobrze on działa. To jest funkcja **strat** lub **błędu**, porównująca przewidywania modelu z prawdziwymi etykietami.

Po obliczeniu błędu musimy również zdefiniować, jak model powinien reagować na zadaną mu informację zwrotną. 
**Algorytm dokonujący optymalizacji wag określa, w jaki sposób sieć uczy się (poprawia/aktualizuje wagi) na podstawie owej opinii.**

```python
loss_function = nn.CrossEntropyLoss()              # najbardziej rozpowszechniona funkcja straty
optimizer = optim.SGD(model.parameters(), lr=0.1)  # algorytm dokonujący optymalizacji wag
```

Połączmy wszystko razem i wytrenujmy naszą sieć!

```python
loss, acc = train_model(model, dataloaders, loss_function, optimizer, num_epochs=3)
print(loss, acc)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16,5))
ax[0].plot(loss)
ax[0].set_title("Błąd")
ax[1].plot(acc)
ax[1].set_title("Dokładność")
```

## Testowanie - sprawdzamy jak skuteczne jest nasze nauczanie

```python
from glob import glob
import math

# zbierz wszystkie obrazy razem i przepuść przez model
validation_img_paths = glob("data/validation/**/*.jpg", recursive=True)
images = [Image.open(img_path) for img_path in validation_img_paths]
validation_batch = torch.stack( [validation_transforms(img) for img in images])
pred_logits_tensor = model(validation_batch)
pred_probs = pred_logits_tensor.cpu().data.numpy()

# pokaż z jaką pewnością sieć przewiduje klasy
fig, axs = plt.subplots(6, 5, figsize=(20, 20))
for i, img in enumerate(images):
    ax = axs[math.floor(i/5)][i % 5]
    ax.axis('off')
    ax.set_title("{:.0f}% Pies, {:.0f}% Ciasto".format(100*pred_probs[i,0], 100*pred_probs[i,1]), fontsize=18)
    ax.imshow(img)
```

```python
# klasy w zbiorze danych COCO
CLASSES = [
    'N/A', 'person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus',
    'train', 'truck', 'boat', 'traffic light', 'fire hydrant', 'N/A',
    'stop sign', 'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse',
    'sheep', 'cow', 'elephant', 'bear', 'zebra', 'giraffe', 'N/A', 'backpack',
    'umbrella', 'N/A', 'N/A', 'handbag', 'tie', 'suitcase', 'frisbee', 'skis',
    'snowboard', 'sports ball', 'kite', 'baseball bat', 'baseball glove',
    'skateboard', 'surfboard', 'tennis racket', 'bottle', 'N/A', 'wine glass',
    'cup', 'fork', 'knife', 'spoon', 'bowl', 'banana', 'apple', 'sandwich',
    'orange', 'broccoli', 'carrot', 'hot dog', 'pizza', 'donut', 'cake',
    'chair', 'couch', 'potted plant', 'bed', 'N/A', 'dining table', 'N/A',
    'N/A', 'toilet', 'N/A', 'tv', 'laptop', 'mouse', 'remote', 'keyboard',
    'cell phone', 'microwave', 'oven', 'toaster', 'sink', 'refrigerator', 'N/A',
    'book', 'clock', 'vase', 'scissors', 'teddy bear', 'hair drier',
    'toothbrush'
]

# kolory do wizualizacji
COLORS = [[0.000, 0.447, 0.741], [0.850, 0.325, 0.098], [0.929, 0.694, 0.125],
          [0.494, 0.184, 0.556], [0.466, 0.674, 0.188], [0.301, 0.745, 0.933]]
```

```python
# standaryzacja dla danych COCO
transform = T.Compose([
    T.Resize(800),
    T.ToTensor(),
    T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# funkcje do przelicenia położeń prostokątów ograniczających dla znalezionych obiektów
def box_cxcywh_to_xyxy(x):
    x_c, y_c, w, h = x.unbind(1)
    b = [(x_c - 0.5 * w), (y_c - 0.5 * h),
         (x_c + 0.5 * w), (y_c + 0.5 * h)]
    return torch.stack(b, dim=1)

def rescale_bboxes(out_bbox, size):
    img_w, img_h = size
    b = box_cxcywh_to_xyxy(out_bbox)
    b = b * torch.tensor([img_w, img_h, img_w, img_h], dtype=torch.float32)
    return b
```

```python
def plot_results(pil_img, prob, boxes):
    plt.figure(figsize=(16,10))
    plt.imshow(pil_img)
    ax = plt.gca()
    colors = COLORS * 100
    for p, (xmin, ymin, xmax, ymax), c in zip(prob, boxes.tolist(), colors):
        ax.add_patch(plt.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin,
                                   fill=False, color=c, linewidth=3))
        cl = p.argmax()
        text = f'{CLASSES[cl]}: {p[cl]:0.2f}'
        ax.text(xmin, ymin, text, fontsize=15,
                bbox=dict(facecolor='yellow', alpha=0.5))
    plt.axis('off')
    plt.show()
```

Pobierzemy z Internetu model detektora z przetrenowanymi już wagami.

```python
model = torch.hub.load('facebookresearch/detr', 'detr_resnet50', pretrained=True)
#model.eval();
```

W tym miejscu pobierzemy obraz z bazy COCO (może uda Ci się wybrać inny ciekawy obrazek?)

```python
# wczytanie obrazu z konkretnej strony internetowej
url = 'http://images.cocodataset.org/val2017/000000039769.jpg'
im = Image.open(requests.get(url, stream=True).raw)

# wczytanie obrazu z dysku
# im = Image.open('./obrazy/000000039769.jpg')
```

```python
# normalizacja obrazu
img = transform(im).unsqueeze(0)

# przejście przez sieć
outputs = model(img)

# zostawiamy jedynie predykcje, których sieć była pewna przynajmniej na 90%
probas = outputs['pred_logits'].softmax(-1)[0, :, :-1]
keep = probas.max(-1).values >= 0.9

# wyliczamy położenie obiektów w pixelach
bboxes_scaled = rescale_bboxes(outputs['pred_boxes'][0, keep], im.size)
```

```python
plot_results(im, probas[keep], bboxes_scaled)
```
